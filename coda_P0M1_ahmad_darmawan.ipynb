### Import Libaries

````python
# data manipulation or dataframe
import pandas as pd

# webscraping
from selenium import webdriver
from bs4 import BeautifulSoup
````

## Steps

### 1. Assign WebDriver

````python
from selenium.webdriver.edge.service import Service
service = Service(r"C:\Users\ASUS\OneDrive\Desktop\web scraping\msedgedriver.exe")
driver = webdriver.Edge(service=service)
````

### 2. Access Website

````python
# link
url = 'https://www.ikea.co.id/in/penawaran/produk-terlaris'
#url = 'https://www.masarishop.com/sale.html?cat=3754'

# access the website
driver.get(url)
````

### 3. Fetch HTML

````python
# fetch html components
html = driver.page_source
````

### 4. Parse HTML Components

````python
# parse
soup = BeautifulSoup(html, "html.parser")
````

### A. Extract 

#### Nama Produk

```html
<span class="itemFacts font-weight-normal">Selimut kecil, abu-abu, 120x160 cm</span>
```

````python
# get element
el_names = soup.find_all('span', {"class", "itemFacts font-weight-normal"})

# empty list
names = []

for name in el_names:
    names.append(name.get_text())

names
````

#### Harga Produk

```html
<span class="config_price diff minimum-show special-price">IDR 9,870,000</span>
<span class="<span class="old-price>IDR 14,100,000</span>">IDR 14,100,000</span>
<span data-price="49900"><span class="currency">Rp</span>49.900</span>
```

````python
# get element
el_price = soup.find_all("span", {"data-price": True})

# empty list
prices = []

for price in el_price:
    price_value = price.get('data-price')  # Ambil nilai dari atribut
    prices.append(price_value)

prices
````

#### Nama Toko

```html
<a class="itemName" href="/in/produk/bantal-kursi-dan-selimut-kecil/selimut-kecil/vitmossa-art-70304890">VITMOSSA</a>
```

````python
# get element
el_toko = soup.find_all("a", {"class":"itemName"})

# empty list
nama_toko = []

for kota in el_toko:
    nama_toko.append(kota.get_text())

nama_toko
````

#### Keterangan 

```html
<div class="sticker_message" style="color: rgb(255, 255, 255);">Produk online terlaris</div>
```

````python
# get element
el_keterangan = soup.find_all('div', {'class':'sticker_message'})

# empty list
semua_keterangan = []

for keterangan in el_keterangan:
    semua_keterangan.append(keterangan.get_text())

semua_keterangan
````

````python
# mematikan/mengeluarkan driver
driver.quit()
````

#### Create Into a Dataframe

````python
data = pd.DataFrame(
    {
        'nama produk':names,
        "harga produk": prices,
        "nama toko" : nama_toko,
        "keterangan" : keterangan
    }
)
data
````
